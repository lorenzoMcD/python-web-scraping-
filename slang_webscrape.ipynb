```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.netlingo.com/acronyms.php'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
```

```python
results = soup.find('div',attrs = {'class': 'list_box3'})
```

```python
li = results.find_all('li')
explain = results.find_all('explain')
```

```python
li
```

```python
explain = results.find_all('div',attrs = {'class': 'explain'})
```

```python
explain
```

```python
meaning = []
for items in explain:
    meaning.append(items.text)
```

```python

lookup = []
for item in li:
    lookup.append(item.span.text)
    
 
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'abrv': lookup,
     'definiton': meaning
    })
```

```python
df.to_csv('txtabrv.csv',index = False, encoding = 'utf-8')

#df = pd.DataFrame(txtabrv, columns=['abrv','defintion',]) # links/sets column names and data
```
