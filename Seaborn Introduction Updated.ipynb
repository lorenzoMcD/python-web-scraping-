```python
df = pd.read_csv('Pokemon.csv')
df.head()
```

```python
# plot a comparison between different types of generation 1 Pokemon
# Generation 1 :  number 1 - 151
#plt.figure(figsize=(10,10))
#df[ df['number'] <= 151]['type1'].value_counts().plot.pie()
df[ df['number'] <= 151]['type1'].value_counts().plot.barh()
#df[df['number'] < 152]['type1'].value_counts().plot(kind='pie')
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

fig.suptitle('Comparison Between Gen1 and Gen2', fontsize = 16, y=1.01)
ax[0].set_title("Generation 1 Type Breakdown")
ax[0].set_xlabel("Count")
ax[0].set_ylabel("Type")
df[df['number'] <= 151]['type1'].value_counts().plot(kind='barh', ax=ax[0])  # , ax=ax[0]
ax[1].set_title("Generation 2 Type Breakdown")
ax[1].set_xlabel("Count")
ax[1].set_ylabel("Type")
df[(df['number'] > 151) & (df['number'] <= 251)]['type1'].value_counts().plot(kind='barh', ax=ax[1])  # , ax=ax[1]
#plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(10,5))
# sns.barplot( x = x_data, y = y_data )
sns.barplot(x=df[df['number'] < 152]['type1'].value_counts(), y=df[df['number'] < 152]['type1'].value_counts().keys())
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
fig.suptitle("Comparison of Gen 1 and Gen 2 Pokemon Types", fontsize=16) # ,y=1.05

ax[0].set_title("Generation 1")
sns.barplot(x=df[df['number'] < 152]['type1'].value_counts(), y=df[df['number'] < 152]['type1'].value_counts().keys(), ax=ax[0])
ax[0].set_xlabel("Count")
ax[0].set_ylabel("Type")

sns.barplot(x=df[(df['number'] >=151) & (df['number'] <= 251)]['type1'].value_counts(), y=df[(df['number'] >=151) & (df['number'] <= 251)]['type1'].value_counts().keys(), ax=ax[1])
ax[1].set_title("Generation 2")
ax[1].set_xlabel("Count")
ax[1].set_ylabel("Type")
```
