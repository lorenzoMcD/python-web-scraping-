```python
import requests
from bs4 import BeautifulSoup
import time    # time.sleep(5) -- this will pause the program for 5 seconds
```

```python
# print the titles of ALL the books on the books.toscrape.com site
count = 1
for i in range(1,51):
    url = 'http://books.toscrape.com/catalogue/page-%d.html' %(i)
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    articles = soup.find_all('article')
    for article in articles:
        print(count, article.img['alt'])
        count += 1
    # add a small time delay to play nicely...
    time.sleep(3)  
    
    
```

```python
# do the same as above using the 'Next' link...
# find average price of all books on this site
priceSum = 0
priceCount = 0

baseurl = 'http://books.toscrape.com/'
url = baseurl
crawl = True

while crawl == True:
    # navigate to url
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup( page.content, 'html.parser')
    
    # scrape title data
    # scrape price value; sum and count it
    price_tags = soup.find_all('p', {'class': 'price_color'})
    for p in price_tags:
        price = float(p.text[1:]) # skip first character in text of this p tag
        priceSum += price
        priceCount += 1
    
    # check for 'Next' button
    nextTag = soup.find('li', {'class':'next'})
    
    # if 'Next' button exists, update url
    if nextTag:
        # Note: starting on page 2, the href format changed, so we
        # need to modify our base url to reflect this change...
        if '3' in nextTag.a['href']:
            baseurl = 'http://books.toscrape.com/catalogue/'
        url = baseurl + nextTag.a['href']
    # else: crawl = False
    else:
        crawl = False

print("Done.")
print("Average price is %.2f" %(priceSum / priceCount))
```
